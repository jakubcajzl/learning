```python
def my_argmax(arr):
    argmax = arr.argmax()
    shape = arr.shape
    loc = np.ndarray(arr.ndim, dtype=int)
    i = arr.ndim - 1
    while i > 0:
        idx = argmax // sum(shape[0:i])
        loc[i] = idx
        argmax -= idx * sum(shape[0:i])
        i -= 1
    
    loc[0] = argmax
    
    return loc

def my_argmax2(arr):
    return np.array(np.unravel_index(arr.argmax(), shape=arr.shape))
```

## Introduction to **Pandas**

### Data types in Pandas

In Pandas, data types are with command `dtype`:

```python
# Data type of a column:
DataFrame['column'].dtype()

# Data type of all columns of a DataFrame:
DataFrame.dtypes()
```

There are **4 main data types** in **Pandas**:
1. `object` = a **text** (similar to a string)
2. `int64` = **integer**
3. `float64` = **floating number**
4. `bool` = **booleans** (True/False)

### Boolean operators in Pandas

In **Pandas** are 3 main `bit-wise` **boolean operators**:
- `and` = **`&`**
- `or` = **`|`**
- `not` = **`~`**

`Bit-wise` boolean operators work **element-wise** (element-by-element) on a **series** (or iterable, in general).

#### Series
- Series = one-dimensional labeled array capable of holding any data type
- Axis labels are reffered to as the **index**
- Index doesn't have to be unique - methods that require unique index will throw error in case of non-unique index

```python
s = pd.Series(['a', 'b', 'c'])
s
```

```python
# if dictionary is passed to Series constructor, dictionary key is used as an index
s = pd.Series({'a': 1, 'b': 2})
s
```

Pandas DataFrame supports categorical data type. It has following features:

- all values are either in `categories` or `np.nan`

```python
data = pd.read_csv('data/TARAN_data_digi_scoring_all.txt', sep = '\t', encoding = 'utf-8', low_memory=False)

dt_grp = data.groupby('Client_Type').agg(
    avg_income = ('HH_Income', 'mean'),
    count = ('CPS_ID', len),
    def_rt = ('DEFAULT_30D_3M', 'mean'),    
)

dt_grp
```

```python
s = dt_grp.style.format(
    na_rep='MISSING',
    thousands=" ",
    precision=0,
    formatter={'def_rt': '{:.2%}'}
)
# .hide_index().hide_columns(['count'])

dt_grp.columns.name=''
```

```python
# Modify headers
cell_hover = {  # for row hover use <tr> instead of <td>
    'selector': 'td:hover',
    'props': [('background-color', '#ffffb3')]
}
index_names = {
    'selector': '.index_name',
    'props': 'font-style: italic; color: darkgrey; font-weight:normal;'
}
headers = {
    'selector': 'th:not(.index_name)',
    'props': 'background-color: #000066; color: white;'
}
s.set_table_styles([cell_hover, index_names, headers])

# Modify header some more
s.set_table_styles([
    {'selector': 'th.col_heading', 'props': 'text-align: center;'},
    {'selector': 'th.col_heading.level0', 'props': 'font-size: 1.2em;'},
    {'selector': 'td', 'props': 'text-align: center; font-weight: bold;'},
], overwrite=False)

s.set_table_styles({
    ('Regression', 'Tumour'): [{'selector': 'th', 'props': 'border-left: 1px solid white'},
                               {'selector': 'td', 'props': 'border-left: 1px solid #000066'}]
}, overwrite=False, axis=0)

# Add highlighting on hover
s.set_table_styles({
    'count': [{'selector': 'th', 'props': 'border-right: 1px solid white'},
              {'selector': 'td', 'props': 'border-right: 1px solid #000066'}]
}, overwrite=False, axis=0)

# Set cell colors
s.set_table_styles({
    'avg_income': [{'selector': '.true', 'props': 'background-color: #e6ffe6;'},
                   {'selector': '.false', 'props': 'background-color: #ffe6e6;'}],
    'count': [{'selector': '.true', 'props': 'background-color: #e6ffe6;'},
              {'selector': '.false', 'props': 'background-color: #ffe6e6;'}],
}, overwrite=False)
cell_color = dt_grp.applymap(lambda x: 'true' if round(x)%2 == 0 else 'false')
s.set_td_classes(cell_color)

# apply style by function
def style_dr(v):
    mid = 0.0021
    if v < mid:
        st = 'color:green;'
    else:
        st = 'color:red;'

    opacity = max(min(abs(v - mid) * 100/ 0.02, 100),40)
    st += f' opacity: {opacity}%;'
    return st

s.applymap(style_dr, subset=['def_rt'])

# set caption
s.set_caption("Some arbitrary table.").set_table_styles([{
    'selector': 'caption',
    'props': 'caption-side: bottom; font-size:1.25em;'
 }], overwrite=False)

# adding tooltip
tooltip = pd.DataFrame([['New to market is by far the worst population']],
                  index=['NTM'], columns=['def_rt'])
s.set_tooltips(
    tooltip, 
    props='visibility: hidden; position: absolute; z-index: 1; border: 1px solid #000066;'
          'background-color: white; color: #000066; font-size: 0.8em;'
          'transform: translate(0px, -24px); padding: 0.6em; border-radius: 0.5em;'
)
```

### Comparing NumPy and DataFrame performance:

#### **Example 01**: sorting

```python
# NumPy array:
a = np.array([6,1,8,6,2,7,6,9])
a
```

```python
# sort in descending order:
%timeit np.sort(a)[::-1]
```

```python
# alternative descending sorting:
%timeit np.flip(np.sort(a))
```

```python
# comparing with pd.dataframe:
b = pd.DataFrame(a, columns=['a'])
b
```

```python
%timeit b.sort_values(by=['a'], ascending=False)
```

```python
149/2.56
```

**Result:** NumPy sorts an np.array in descending order approx. **58-times faster** than Pandas pd.DataFrame
